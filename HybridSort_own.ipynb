## Random Array Generator

```python
def random_array_generator(listSize):
    randomList = np.random.randint(low=1, high=listSize, size=listSize)
    random.shuffle(randomList)
    return randomList

def random_array_generator_old(listSize, maxCase):
    randomList = []
    for i in range(listSize):
        randomList.append(random.randint(0, maxCase))
    return randomList
```

## Merge Sort

```python
def merge_sort(mainList):
    keyComparisons = 0
    
    mid = len(mainList)//2 # To get middle of list
    # Split list into 2 halves
    leftList = mainList[:mid]
    rightList = mainList[mid:]
    
    if mid <= 0: # Base Case, check if only 1 or 0 element in list
        return 0
    else: # Split list into 2 lists, left and right
        keyComparisons += merge_sort(leftList) # Recursive
        keyComparisons += merge_sort(rightList) # Recursive
        
        # Merging of 2 sorted lists together into 1 list
        mainListPointer = 0
        leftListPointer = 0 # Points to first element of first list
        rightListPointer = 0 # Points to second element of second list

        while (leftListPointer < len(leftList)) and (rightListPointer < len(rightList)): # While both lists are not empty
            if leftList[leftListPointer] < rightList[rightListPointer]: # If 1st element of left half is smaller
                mainList[mainListPointer] = leftList[leftListPointer] # Append 1st element of left half to main merged list
                mainListPointer += 1
                leftListPointer += 1
            else: # If 1st element of right half is smaller
                mainList[mainListPointer] = rightList[rightListPointer] # Append 1st element of right half to main merged list
                mainListPointer += 1
                rightListPointer += 1

            keyComparisons += 1

        # Merge any remaining items from leftList or rightList into mainList
        while (leftListPointer < len(leftList)): # Merge if there remain items in leftList
            mainList[mainListPointer] = leftList[leftListPointer]
            mainListPointer += 1
            leftListPointer += 1

        while (rightListPointer < len(rightList)): # Merge if there remain items in rightList
            mainList[mainListPointer] = rightList[rightListPointer]
            mainListPointer += 1
            rightListPointer += 1
            
        return keyComparisons
```

```python
def insertion_sort(mainList):

    # Traverse through 1 to len(arr)
    comp = 0
    for i in range(1, len(mainList)):
        key = mainList[i]
  
        # Move elements of arr[0..i-1], that are
        # greater than key, to one position ahead
        # of their current position
        j = i-1
        while j >= 0 and key < mainList[j] :
                mainList[j + 1] = mainList[j]
                j -= 1
                comp +=1
        mainList[j + 1] = key
    return comp
```

## Hybrid Sort

##### Setting a small integer S as a threshold. Once the size of a subarray in a recursive call of Merge Sort is less than or equal to S, algorithm will switch to Insertion Sort, efficient for small-sized input.

```python
def hybrid_sort(mainList, S):
    keyComparisons = 0
    
    if len(mainList) <= S:
        # Insertion Sort
    
        if (len(mainList) > 1):
            # Assuming (len(mainList) >= 1); ommiting 2 cases where list has only 0 or 1 element which is sorted
            for i in range(len(mainList)): # Outer loop, pick a new item from mainList
                for j in range(i, 0, -1): # Inner loop, finding the correct posiiton to insert the item
                    keyComparisons += 1
                    
                    if (mainList[j] < mainList[j-1]): # Comparing between the 'sorted' & 'unsorted' sections of list
                        # If item in 'unsorted' section smaller, swap
                        temp = mainList[j]
                        mainList[j] = mainList[j-1]
                        mainList[j-1] = temp
                    else:
                        break;
        else:
            return 0
                    
    else:
        # Merge Sort
        
        mid = len(mainList)//2 # To get middle of list
        # Split list into 2 halves
        leftList = mainList[:mid]
        rightList = mainList[mid:]

        # Base Case, check if only 1 or 0 element in list
        if mid <= 0:
            return 0
        else:
            keyComparisons += hybrid_sort(leftList, S) # Recursive
            keyComparisons += hybrid_sort(rightList, S) # Recursive

            # Merging of 2 sorted lists together into 1 list
            mainListPointer = 0
            leftListPointer = 0 # Points to first element of first list
            rightListPointer = 0 # Points to second element of second list

            while (leftListPointer < len(leftList)) and (rightListPointer < len(rightList)): # While both lists are not empty
                if leftList[leftListPointer] < rightList[rightListPointer]: # If 1st element of left half is smaller
                    mainList[mainListPointer] = leftList[leftListPointer] # Append 1st element of left half to main merged list
                    mainListPointer += 1
                    leftListPointer += 1
                else: # If 1st element of right half is smaller
                    mainList[mainListPointer] = rightList[rightListPointer] # Append 1st element of right half to main merged list
                    mainListPointer += 1
                    rightListPointer += 1
                    
                keyComparisons += 1

            # Merge any remaining items from leftList or rightList into mainList
            while (leftListPointer < len(leftList)): # Merge if there remain items in leftList
                mainList[mainListPointer] = leftList[leftListPointer]
                mainListPointer += 1
                leftListPointer += 1

            while (rightListPointer < len(rightList)): # Merge if there remain items in rightList
                mainList[mainListPointer] = rightList[rightListPointer]
                mainListPointer += 1
                rightListPointer += 1
                
    return keyComparisons
```
